```python
def getshape(dataset):
    return (pd.read_csv(dataset).shape)
def getduration(dataset):
    df = pd.read_csv(dataset)
    df.date = pd.to_datetime(df.date)
    return(str(df.date.min()), str(df.date.max()), str(df.date.diff()[1]))
```

```python
import os



for subdir, dirs, files in os.walk("LTSF_Linear/LTSF-Linear/dataset"):
    for file in files:
        #print os.path.join(subdir, file)
        filepath = subdir + os.sep + file
        if filepath.endswith(".csv"):
            print (filepath.split("/")[-1],": ", getshape(filepath), getduration(filepath))
            
```
